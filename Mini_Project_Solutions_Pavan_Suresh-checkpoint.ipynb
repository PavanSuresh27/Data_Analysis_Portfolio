```python
conversion = {
    "age": int,"job":str,"marital":str,"education":str,"default":str,"balance":float,"housing":str,"loan":str,"contact":str,
    "day":float,"month":str,"duration":float,"campaign":int,"pdays":int,"previous":int,"poutcome":str,"y":str
}
def convert_data_types(data_frame, conversion):
    for col, data_type in conversion.items():
        data_frame[col] = data_frame[col].astype(data_type)
    return data_frame
df1_converted = convert_data_types(df1.copy(), conversion)
updated_data_types = df1_converted.dtypes
print(updated_data_types)
```

# Find the average balance of the customer who belongs to the subscribed customer and non-subscribed customer and also use a related plot to show them in visualization.

```python
subscribed = df1[df1['y'] == 'yes']['balance'].mean()

non_subscribed = df1[df1['y'] == 'no']['balance'].mean()

labels = ['Subscribed', 'Non-Subscribed']
avg_balances = [subscribed,subscribed]

plt.bar(labels, avg_balances, color=['red', 'yellow'])
plt.xlabel('Customer Category')
plt.ylabel('Average Balance')
plt.title('Average Balance by Customer Subscription')
plt.show()


print("Average Balance Subscribed:",subscribed)
print("Average Balance - Non-Subscribed:",non_subscribed)
```
